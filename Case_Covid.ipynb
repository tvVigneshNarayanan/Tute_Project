```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset  = pd.read_csv("/content/covid19_Confirmed_dataset.csv")
dataset.head()
```

```python
df = dataset.drop(["Lat","Long"],axis=1,inplace=True)
```

```python
covid_data_aggregate = dataset.groupby("Country/Region").sum()
```

```python
countries = list(covid_data_aggregate.index)
maximum_infected_rates = []
for co in countries:
  maximum_infected_rates.append(covid_data_aggregate.loc[co].diff().max())
covid_data_aggregate["Maximum Infection Rates"]=  maximum_infected_rates
```

```python
covid_data_aggregate
```

```python
Covid_data = pd.DataFrame(covid_data_aggregate["Maximum Infection Rates"])
```

```python
Covid_data
```
